```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Assuming you have your data in a CSV file, replace 'your_dataset.csv' with the actual filename
df = pd.read_csv('farm_sprayers.csv')

# Remove '%' and 'off' from the 'Offers' column and convert it to float
df['Offers'] = df['Offers'].str.replace(r'%\s*off', '', regex=True).astype(float)

# Prepare the data for PCA
X_pca = df[['Product Price', 'Offers']]

# Perform PCA to combine 'Product Price' and 'Offers' into one column
pca = PCA(n_components=1)
X_pca['PCA'] = pca.fit_transform(X_pca)

# Combine PCA column with 'Product Rating'
X_combined = pd.concat([X_pca['PCA'], df['Product Rating']], axis=1)

# Create a scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(X_combined['PCA'], X_combined['Product Rating'], alpha=0.7)

plt.xlabel('PCA (Product Price + Offers)')
plt.ylabel('Product Rating')
plt.title('Product Data Scatter Plot with PCA')
# plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming you have your data in a CSV file, replace 'your_dataset.csv' with the actual filename
df = pd.read_csv('farm_sprayers.csv')

# Remove '%' and 'off' from the 'Offers' column and convert it to float
df['Offers'] = df['Offers'].str.replace(r'%\s*off', '', regex=True).astype(float)

# Prepare the data for PCA
X = df[['Product Price', 'Offers']]

# Perform PCA to combine 'Product Price' and 'Offers' into one column
pca = PCA(n_components=1)
X['PCA'] = pca.fit_transform(X)

# Perform K-means clustering on the combined data (PCA, Rating)
num_clusters = 3  # Best, Good, Bad
kmeans_model = KMeans(n_clusters=num_clusters, random_state=42)
kmeans_model.fit(X_combined)

# Add the cluster labels to the DataFrame
df['Cluster'] = kmeans_model.labels_

# new_data = pd.concat([X_combined],[df['Cluster']], axis=1)
plt.figure(figsize=(10,10))
plt.scatter(X_combined['PCA'], X_combined['Product Rating'],c=df['Cluster'],cmap='plasma')
plt.xlabel('PCA (Product Price + Offers)')
plt.ylabel('Product Rating')
plt.title('Product Data Scatter Plot with PCA')
# plt.grid(True)
plt.show()
```

```python
# Create a dictionary to store the product names for each label category
product_names_by_label = {}

# Iterate through the dataset and group product names by label category
for label, product_name in zip(df['Cluster'], df['Product Name']):
    if label not in product_names_by_label:
        product_names_by_label[label] = [product_name]
    else:
        product_names_by_label[label].append(product_name)

# Print the separated product names for each label category
for label, product_names in product_names_by_label.items():
    print(f"Label '{label}':")
    for product_name in product_names:
        print(f"  - {product_name}")
    print()  # Print an empty line between label categories
```
